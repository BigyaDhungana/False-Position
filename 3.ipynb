## Finite Difference Table

```python
import numpy as np
import pandas as pd
import math
```

```python
f=lambda x:(math.e)**x
start=-1
end=1
step=0.1
precision=6
```

```python

x_array=np.round(np.arange(start,end+step,step),1)
y_array=np.round(f(x_array),precision)
```

```python
data={
    'x':x_array,
    'y':y_array,
}
```

```python
def difference(array,precision,index):
    diff_array=[""]*len(array)
    j=0
    for i in range(1,len(array)-index):
        diff=round(array[i]-array[j],precision)
        diff_array[j]=diff
        j=j+1
    return diff_array
```

```python
obt_list=y_array.tolist()
for i in range(0,len(y_array)-1):
    obt_list=difference(obt_list,precision,i)
    data["{index}".format(index=i)]=obt_list
```

```python
data_frame=pd.DataFrame(data)
print(data_frame)
```
